```python
from numpy import sqrt, sum, histogram
from matplotlib.pyplot import bar

def CLT (dist, n, repeat):
    R= (dist.rvs((n, repeat)) - dist.expect()) / sqrt(n) / dist.std()
    hist, bins = histogram(sum(R,axis=0), bins="auto")
    bar(bins[:-1], hist / sum(hist))
```

rvs (random variants)
- zufällige n Werte gemäss der vorgegebenen Vertielung

expect
- subtrahiere "müh" (Erwartungswert). expect liefertt von der Verteilung den Erwartungswert

std (standard diviation)
